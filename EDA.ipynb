```python
import os
import re
import bs4 
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
pd.set_option("display.max_rows", None, "display.max_columns", 10)
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("after_data_handling.csv",index_col=[0],thousands=',')
```

```python
df.info()
```

```python
price_Fuel=pd.crosstab(df["Fuel type"],df["bind_price"])
```

```python
price_Fuel.plot(kind='bar',figsize=(10,5))
plt.title('Fuel type vs. price',fontsize=10)
plt.xlabel('Fuel type')
plt.xlabel('price')
```
